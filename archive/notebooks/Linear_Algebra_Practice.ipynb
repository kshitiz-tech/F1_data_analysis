```python
iris = pd.read_csv("iris.csv",sep = "\t")
#take the versicolor species renaming the columns to short version
versicolor = iris[iris.Species == "versicolor"].rename(
    columns = {"Petal.Length" : "plength", "Petal.Width" :"pwidth"})
versicolor.head(3)
```

```python
plt.scatter(versicolor.plength, versicolor.pwidth)
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
```

```python
def mse(beta0, beta1):
    pred_width  = beta0 + beta1*versicolor.plength
    e = versicolor.pwidth - pred_width
    mse = np.mean(e**2)
    return mse
```

```python
mse(0,1)
```

```python
mse(0,0.5)
```

```python
mse(-0.5,0.5)
```

```python
beta0, beta1 = -0.5, 0.5
plt.scatter(versicolor.plength, versicolor.pwidth, edgecolor = "k")

#creating 11 evenly spaced length
length = np.linspace(versicolor.plength.min(), versicolor.plength.max(),11)

#predicting width
predw = beta0 + beta1*length
plt.plot(length, predw, c = 'r')
plt.xlabel("Petal length")
plt.ylabel("Petal width")
```

```python
import statsmodels.formula.api as smf

m = smf.ols("pwidth ~ plength", data = versicolor).fit()

```
